```python


Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.
ans-
The Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical functions that are used to describe the probability distribution of a random variable.

The Probability Mass Function (PMF) is used for discrete random variables, which take on a finite or countable number of values. The PMF gives the probability of each value that the random variable can take.

For example, consider a six-sided die. The possible outcomes are {1, 2, 3, 4, 5, 6}. The PMF for this random variable gives the probability of each of these outcomes. Since each outcome is equally likely, the PMF is given by:

P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6

where X is the random variable representing the outcome of rolling the die.

The Probability Density Function (PDF) is used for continuous random variables, which can take on any value within a certain range. The PDF gives the probability of the random variable falling within a certain range of values.

For example, consider the height of adult men in a certain population. The height can take on any value between 0 and infinity (although in practice it is bounded). The PDF for this random variable gives the probability of a man having a height between two values, say a and b. The PDF is often denoted by f(x), where x is the value of the random variable.

Note that the PDF itself does not give the probability of a specific value of the random variable, but rather the probability density (i.e., the probability per unit of value). The probability of the random variable falling within a certain range can be found by integrating the PDF over that range.

For example, suppose the PDF for the height of adult men is given by:

f(x) = 0.02 * e^(-0.02*x)

where x is the height in inches. The probability of a man having a height between 68 and 70 inches is given by:

P(68 ≤ X ≤ 70) = ∫68 to 70 f(x) dx = ∫68 to 70 0.02 * e^(-0.02*x) dx

which can be evaluated to give the probability.




```

```python
Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.
ans-
The normal distribution, also known as the Gaussian distribution, is a commonly used probability distribution in statistics and probability theory. It is often used as a model for random variables in situations where the data tends to cluster around a central value and is symmetrically distributed around that value.

Some examples of situations where the normal distribution might be used as a model include:

Heights of adult humans: Heights of adult humans tend to cluster around a central value, and the distribution of heights is approximately symmetric.

IQ scores: IQ scores are often modeled using the normal distribution, as they tend to be normally distributed in the population.

Errors in measurements: Errors in measurements of physical quantities, such as length or weight, are often modeled using the normal distribution.

The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ). The mean represents the central value around which the data tends to cluster, while the standard deviation represents the spread of the data around the mean.

The shape of the normal distribution is bell-shaped and symmetric around the mean. The standard deviation determines the width of the bell curve: a larger standard deviation leads to a wider bell curve, while a smaller standard deviation leads to a narrower bell curve. The mean determines the position of the peak of the bell curve: if the mean is larger, the peak will be shifted to the right, while if the mean is smaller, the peak will be shifted to the left.

The normal distribution is often used as a model for real-world data because of the central limit theorem, which states that the sum or average of a large number of independent random variables tends to be normally distributed, regardless of the distribution of the individual variables. This makes the normal distribution a useful model in many situations where the data can be treated as a sum or average of underlying random variables.




```

```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
ans-
Bernoulli distribution is a discrete probability distribution that describes the outcome of a single binary (two possible outcomes) experiment, where success is defined as the occurrence of one specific event and failure is defined as the non-occurrence of that event. The distribution is named after the Swiss mathematician Jacob Bernoulli.

The Bernoulli distribution has only one parameter, p, which represents the probability of success in a single trial. The probability mass function (PMF) of a Bernoulli distribution is given by:

P(X = x) = p^x (1-p)^(1-x)

where X is a random variable that takes on values of 0 or 1 (failure or success, respectively).

An example of a Bernoulli distribution is the probability of getting a head when flipping a fair coin. In this case, p=0.5, since the probability of getting a head (success) is 0.5, and the probability of getting a tail (failure) is also 0.5. Another example is the probability of a person having blue eyes, where success is having blue eyes and failure is having any other eye color.

The main difference between the Bernoulli distribution and the Binomial distribution is that the Bernoulli distribution models a single trial or experiment, while the Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. The Binomial distribution has two parameters: the number of trials (n) and the probability of success in each trial (p).

In other words, the Binomial distribution is the sum of multiple independent and identically distributed Bernoulli trials. Therefore, the Binomial distribution has a larger sample space, including all the possible combinations of successes and failures in n trials. The probability mass function of a Binomial distribution is given by:

P(X = x) = (n choose x) * p^x * (1-p)^(n-x)

where X is the number of successes, n is the number of trials, and p is the probability of success in each trial.




```
